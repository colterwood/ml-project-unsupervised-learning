```python
# Standardize data for testing
# Define monetary spend columns
monetary_columns = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']

# Initialize scaler and apply it
scaler_standard = StandardScaler()
df_standardized = df.copy()
df_standardized[monetary_columns] = scaler_standard.fit_transform(df[monetary_columns])

# Display the first few rows
print(df_standardized.head())

# Convert df and df_standardized to csv
df.to_csv("df.csv",index=False)
df_standardized.to_csv("df_standardized.csv",index=False)
```

```python
# Normalize data for testing
# Initialize MinMaxScaler and apply it
scaler_minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[monetary_columns] = scaler_minmax.fit_transform(df[monetary_columns])

# Display the first few rows
print(df_minmax.head())

# Convert df_normalized to csv
df_minmax.to_csv("df_minmax.csv",index=False)
```

### Because of the models we are working with, we will focus on the df_standardized dataset

```python
# Check for multicollinearity
# Compute correlation matrix
corr_matrix = df_standardized[monetary_columns].corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Find highly correlated feature pairs
threshold = 0.9
high_corr_pairs = [(col1, col2) for col1 in monetary_columns for col2 in monetary_columns
                   if col1 != col2 and abs(corr_matrix.loc[col1, col2]) > threshold]

print("Highly correlated feature pairs (corr > 0.9):", high_corr_pairs)
```

```python
#Feature selection using PCA
# Apply PCA on standardized data
pca = PCA(n_components=len(monetary_columns))  # Keep all components
pca.fit(df_standardized[monetary_columns])

# Get explained variance ratio
explained_variance = pca.explained_variance_ratio_

# Get feature contributions to each component (loading scores)
components = pd.DataFrame(pca.components_, columns=monetary_columns, index=[f'PC{i+1}' for i in range(len(monetary_columns))])

# Display explained variance per component
print("Explained variance per principal component:\n", explained_variance)
print("\nFeature contributions to first few PCs:\n", components.head())
```

```python
# Drop columns that are too highly correlated or not being used in modelling
df_reduced = df_standardized.drop(columns=['Detergents_Paper','Channel','Region'])

# Display the first few rows to verify
print(df_reduced.head())
```

```python
# Define the remaining monetary columns after dropping Detergents_Paper, Channel, and Region
remaining_columns = df_reduced.columns.tolist()

# Apply PCA on df_reduced
pca = PCA(n_components=len(remaining_columns))  # Keep all components
pca.fit(df_reduced)

# Get explained variance ratio
explained_variance = pca.explained_variance_ratio_

# Get feature contributions to each principal component
components = pd.DataFrame(pca.components_, columns=remaining_columns, index=[f'PC{i+1}' for i in range(len(remaining_columns))])

# Display explained variance
print("Explained variance per principal component:\n", explained_variance)

# Display feature contributions to first few principal components
print("\nFeature contributions to first few PCs:\n", components.head())
```

```python
# Identify features contributing significantly to PC1 and PC2
top_features = components.iloc[:2].abs().mean().sort_values(ascending=False).head(4).index.tolist()
print(f"Selected Features: {top_features}")
```

```python
# create final df with just top_features
df_final = df_reduced[top_features]

df_final.head()

# Convert df_final to csv
df_final.to_csv("df_final.csv", index=False)
```

# Part II - KMeans Clustering

The objective of the analysis is to group similar products together into clusters based on their attributes such as fresh, milk, grocery, frozen, detergents_paper, and delicatessen. To perform the k-means clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters, initialize the centroids, assign data points to clusters, update the centroids, and repeat until convergence.

```python
df_final = pd.read_csv("df_final.csv")
```

```python
# Run elbow method to determine optimal k v
# Try different values of k (1 to 10)
inertia = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_final)
    inertia.append(kmeans.inertia_)  # Sum of squared distances to centroids

# Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertia, marker='o', linestyle='-')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (Sum of Squared Distances)")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
# Get Silhouette Score
# Try k = 3, 4, 5, 6
for k in [3, 4, 5, 6]:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(df_final)
    score = silhouette_score(df_final, labels)
    print(f"Silhouette Score for k = {k}: {score:.4f}")
```

### Choosing k=3 due to highest Silhouette Score

```python
# Run KMeans using k=3
kmeans_final = KMeans(n_clusters=3, random_state=42, n_init=10)
df_final['Cluster'] = kmeans_final.fit_predict(df_final)

# Show first few rows with cluster labels
print(df_final.head())
```

```python
# Check cluser sizes
print(df_final["Cluster"].value_counts())
```

```python
# Average feature values per cluster
cluster_means = df_final.groupby("Cluster").mean()
print(cluster_means)
```

```python
# Initialize K-Means with k-means++ centroids
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42, n_init=1)

# Fit K-Means without assigning clusters yet
kmeans.fit(df_final.drop(columns=["Cluster"]))  # Drop existing cluster labels

# Get initialized centroids
initial_centroids = kmeans.cluster_centers_
print("Initialized Centroids:\n", initial_centroids)
```

```python
# Assign each data point to the nearest centroid
df_final["Cluster"] = kmeans.predict(df_final.drop(columns=["Cluster"]))

# Check how many points are in each cluster
print(df_final["Cluster"].value_counts())
```

```python
# Compute new centroids as the mean of all points in each cluster
new_centroids = df_final.groupby("Cluster").mean().values

# Compare old vs. new centroids
print("Updated Centroids:\n", new_centroids)
```

```python
# Compare final cluster means
cluster_means = df_final.groupby("Cluster").mean()
print(cluster_means)
```

```python
# Heatmap to compare clusters
# Compute cluster means
cluster_means = df_final.groupby("Cluster").mean()

# Heatmap of cluster feature values
plt.figure(figsize=(8, 5))
sns.heatmap(cluster_means, annot=True, cmap="coolwarm", center=0, linewidths=0.5)
plt.title("Final Cluster Feature Comparisons (Standardized)")
plt.xlabel("Features")
plt.ylabel("Clusters")
plt.show()
```

```python
# Pairplot to visualize clusters by feature pairs
sns.pairplot(df_final, hue="Cluster", palette="tab10", diag_kind="hist")
plt.suptitle("Final Cluster Comparison Across Features", y=1.02)
plt.show()
```

```python
# Compute Davies-Bouldin score for K-Means clusters
db_score = davies_bouldin_score(df_final.drop(columns=["Cluster"]), df_final["Cluster"])
print(f"Davies-Bouldin Score for K-Means: {db_score:.4f}")
```

```python
# Compute CH score
ch_score = calinski_harabasz_score(df_final.drop(columns=["Cluster"]), df_final["Cluster"])
print(f"Calinski-Harabasz Score for K-Means: {ch_score:.4f}")
```

# Part III - Hierarchical Clustering 

Hierarchical clustering is a popular unsupervised machine learning algorithm that is used to identify patterns and group similar data points together in a hierarchy. The algorithm works by iteratively merging or splitting clusters based on a similarity measure until a dendrogram is formed.

To perform hierarchical clustering analysis, you will need to pre-process the dataset, determine the optimal number of clusters using techniques such as dendrogram.

```python
df_final = pd.read_csv("df_final.csv")
```

```python
# Create a dendrogram
plt.figure(figsize=(10, 6))
sch.dendrogram(sch.linkage(df_final, method="ward"))
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
# Determine Correlation for existing dendogram
# Compute linkage matrix
Z = linkage(df_final, method="ward")

# Compute cophenetic correlation coefficient
coph_corr, _ = cophenet(Z, pdist(df_final))
print(f"Cophenetic Correlation Coefficient: {coph_corr:.4f}")
```

```python
# Try k = 3, 4, 5 and compute silhouette scores
for k in [3, 4, 5]:
    hc = AgglomerativeClustering(n_clusters=k, metric="euclidean", linkage="ward")
    labels = hc.fit_predict(df_final)
    score = silhouette_score(df_final, labels)
    print(f"Silhouette Score for k = {k}: {score:.4f}")
```

```python
# Run Agglomerative Clustering with k = 5
hc_final = AgglomerativeClustering(n_clusters=5, metric="euclidean", linkage="ward")
df_final["Cluster"] = hc_final.fit_predict(df_final)

# Show cluster counts
print(df_final["Cluster"].value_counts())
```

```python
# Replot Dendogram
# Create a new dendrogram with color labels
plt.figure(figsize=(10, 6))
sch.dendrogram(sch.linkage(df_final, method="ward"), truncate_mode="lastp", p=10, show_leaf_counts=True)
plt.title("Dendrogram (Truncated to Last 10 Merges)")
plt.xlabel("Cluster Merges")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
# Compute Silhouette Score for Agglomerative Clustering
silhouette_hc = silhouette_score(df_final.drop(columns=["Cluster"]), df_final["Cluster"])
print(f"Silhouette Score for Hierarchical Clustering: {silhouette_hc:.4f}")
```

```python
# Compute DBI for Hierarchical Clustering
db_score_hc = davies_bouldin_score(df_final.drop(columns=["Cluster"]), df_final["Cluster"])
print(f"Davies-Bouldin Score for Hierarchical Clustering: {db_score_hc:.4f}")
```

```python
# Compute CH Score for Hierarchical Clustering
ch_score_hc = calinski_harabasz_score(df_final.drop(columns=["Cluster"]), df_final["Cluster"])
print(f"Calinski-Harabasz Score for Hierarchical Clustering: {ch_score_hc:.4f}")
```

#### Due to the low Cophenetic Correlation Coefficient: 0.4070 and the lower silhouette scores than KMeans, we prefer KMeans to Hierarchical Clustering

# Part IV - PCA

In this section you are going to perform principal component analysis (PCA) to draw conclusions about the underlying structure of the wholesale customer data. Since using PCA on a dataset calculates the dimensions which best maximize variance, we will find which compound combinations of features best describe customers.

```python
# Run PCA
pca = PCA()
pca.fit(df_final.drop(columns=["Cluster"]))

# Get explained variance for each component
explained_variance = np.cumsum(pca.explained_variance_ratio_)
print("Explained Variance by Component:\n", explained_variance)
```

```python
# Get PCA components (loadings)
loadings = pd.DataFrame(pca.components_, columns=df_final.drop(columns=["Cluster"]).columns, index=[f"PC{i+1}" for i in range(pca.n_components_)])
print(loadings)
```

```python
# Check variance without PC1
explained_variance_no_pc1 = pca.explained_variance_ratio_[1:].sum()
print(f"Explained Variance Without PC1: {explained_variance_no_pc1:.4f}")
```

```python
# Apply PCA with 3 components
pca_final = PCA(n_components=3)
df_pca = pca_final.fit_transform(df_final.drop(columns=["Cluster"]))

# Convert PCA results into a DataFrame
df_pca = pd.DataFrame(df_pca, columns=["PC1", "PC2", "PC3"])

# Add back cluster labels for analysis
df_pca["Cluster"] = df_final["Cluster"]

# Show first few rows
print(df_pca.head())
```

```python
# Create a 3D scatter plot of PCA-transformed data
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="3d")

# Scatter plot using PC1, PC2, PC3
scatter = ax.scatter(df_pca["PC1"], df_pca["PC2"], df_pca["PC3"], c=df_pca["Cluster"], cmap="tab10", alpha=0.7)

# Labels
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
ax.set_title("3D PCA Cluster Visualization")
plt.legend(*scatter.legend_elements(), title="Clusters")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df_pca["PC1"], y=df_pca["PC2"], hue=df_pca["Cluster"], palette="tab10", alpha=0.7)
plt.title("PCA Clusters (PC1 vs. PC2)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# Check cluster sizes
print(df_pca["Cluster"].value_counts())
```

```python
# Run K-Means on PCA-transformed data
kmeans_pca = KMeans(n_clusters=3, random_state=42, n_init=10)
df_pca["Cluster"] = kmeans_pca.fit_predict(df_pca.drop(columns=["Cluster"]))

# Show new cluster sizes
print(df_pca["Cluster"].value_counts())
```

```python
# Create a 3D scatter plot of PCA-transformed data
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="3d")

# Scatter plot using PC1, PC2, PC3
scatter = ax.scatter(df_pca["PC1"], df_pca["PC2"], df_pca["PC3"], c=df_pca["Cluster"], cmap="tab10", alpha=0.7)

# Labels
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
ax.set_title("Final 3D PCA Cluster Visualization")
plt.legend(*scatter.legend_elements(), title="Clusters")
plt.show()
```

```python
# Replot scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(x=df_pca["PC1"], y=df_pca["PC2"], hue=df_pca["Cluster"], palette="tab10", alpha=0.7)
plt.title("Final PCA Clusters (PC1 vs. PC2)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# Compute Silhouette Score
silhouette_pca = silhouette_score(df_pca.drop(columns=["Cluster"]), df_pca["Cluster"])
print(f"Silhouette Score for PCA Clusters: {silhouette_pca:.4f}")
```

```python
# Compute DBI
db_score_pca = davies_bouldin_score(df_pca.drop(columns=["Cluster"]), df_pca["Cluster"])
print(f"Davies-Bouldin Score for PCA Clusters: {db_score_pca:.4f}")
```

```python
# Compute CH Score
ch_score_pca = calinski_harabasz_score(df_pca.drop(columns=["Cluster"]), df_pca["Cluster"])
print(f"Calinski-Harabasz Score for PCA Clusters: {ch_score_pca:.4f}")
```
